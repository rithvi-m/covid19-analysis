```python
# COVID-19 Data Analysis - Kodbud Internship Task 1
# By: Your Name

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('country_wise_latest.csv')
print("Data loaded! Shape:", df.shape)

# Clean data
df = df.dropna()
print("Cleaned! Rows:", len(df))

# Chart 1 - Daily Cases
top10_new = df.nlargest(10, 'New cases')
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_new, x='New cases', y='Country/Region', palette='Blues_r')
plt.title('Daily New COVID-19 Cases - Top 10 Countries')
plt.xlabel('New Cases')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('chart1_daily_cases.png')
plt.show()

# Chart 2 - Deaths
top10_deaths = df.nlargest(10, 'Deaths')
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_deaths, x='Deaths', y='Country/Region', palette='Reds_r')
plt.title('Total COVID-19 Deaths - Top 10 Countries')
plt.xlabel('Deaths')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('chart2_deaths.png')
plt.show()

# Chart 3 - Recovery Trends
region_data = df.groupby('WHO Region')['Recovered / 100 Cases'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=region_data.values, y=region_data.index, palette='Greens_r')
plt.title('COVID-19 Recovery Rate by WHO Region')
plt.xlabel('Recovered per 100 Cases')
plt.ylabel('WHO Region')
plt.tight_layout()
plt.savefig('chart3_recovery.png')
plt.show()
```
